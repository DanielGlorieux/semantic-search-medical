# Exploration des Données
Notebook d'exploration du corpus de documents

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
%matplotlib inline
```

## Chargement des données

```python
# Charger le corpus
docs = pd.read_csv('../data/processed/docs.csv')
print(f"Nombre de documents: {len(docs)}")
docs.head()
```

## Statistiques de base

```python
# Longueur des documents
docs['text_length'] = docs['text'].str.len()
docs['word_count'] = docs['text'].str.split().str.len()

print(docs[['text_length', 'word_count']].describe())
```

## Visualisations

```python
# Distribution de la longueur des documents
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].hist(docs['text_length'], bins=50, edgecolor='black')
axes[0].set_title('Distribution de la longueur du texte')
axes[0].set_xlabel('Longueur (caractères)')
axes[0].set_ylabel('Fréquence')

axes[1].hist(docs['word_count'], bins=50, edgecolor='black', color='orange')
axes[1].set_title('Distribution du nombre de mots')
axes[1].set_xlabel('Nombre de mots')
axes[1].set_ylabel('Fréquence')

plt.tight_layout()
plt.show()
```
